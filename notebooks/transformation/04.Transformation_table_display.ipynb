##### **02.- Eliminar columna environment**

```python
df2 = df1.drop("environment")
```

```python
df_final = df2.fillna({"refresh_rate": "Unknown"})
```

```python
display(df_final.head(10))
```

```python
df_final.printSchema()
```

##### **03.- Escribir datos en lh_gold**

```python
merge_delta_lake(
    table_name="lh_gold.dbo.dim_display",
    table_path="abfss://Smart_device_Analytics_WS@onelake.dfs.fabric.microsoft.com/lh_gold.Lakehouse/Tables/dbo/dim_display",
    merge_condition= "tgt.display_id = src.display_id AND tgt.file_date = src.file_date",
    input_df= df_final,
    partition_column= "file_date"
)
```
