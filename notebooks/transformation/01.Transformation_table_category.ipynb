# **Transformaci√≥n de tabla category**
##### **Ejemplo de una carga completa, no incremental**

##### **01.- Leer tabla con spark.sql**

```python
df1 = spark.sql("SELECT * FROM lh_silver.dbo.category")
```

##### **02.- Eliminar columna environment**

```python
df2 = df1.drop("environment")
```

##### **03.- Escribir datos en lh_gold**

```python
df2.write.format("delta").mode("overwrite").saveAsTable("dim_category")
```

```python
%%sql
select * from dim_category
limit(10)
```
