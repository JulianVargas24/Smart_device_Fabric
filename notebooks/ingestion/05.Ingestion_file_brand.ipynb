```python
brand_schema = "BrandID INT, Brand STRING, HardwareDesigner STRING, \
                Manufacturer STRING"
```

```python
df = spark.read.schema(brand_schema) \
                .json(f"{bronze_folder_path}/{p_file_date}/brand.json")
```

##### **2.- Cambiar nombres de columnas**

```python
new_cols = ["brand_id", "brand", "hardware_designer", "manufacturer"]

df2 = df.toDF(*new_cols)
```

##### **3.- Crear ingestion_date y environment**

```python
from pyspark.sql.functions import lit

df3 = add_ingestion_date(df2) \
        .withColumn("environment", lit(p_environment)) \
        .withColumn("file_date", lit(p_file_date))
```

```python
%%sql
DROP TABLE if EXISTS brand
```

##### **4.- Pasar el archivo a delta y lh_silver**

```python
df3.write.format("delta").mode("overwrite").saveAsTable("brand")
```
