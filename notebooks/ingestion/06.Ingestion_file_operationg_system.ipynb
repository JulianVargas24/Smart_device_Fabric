# **Ingestion del archivo operationg_system.json**

##### **01.- Leer archivo JSON usando dataframereader de spark**

```python
#p_file_date = "2025-07-07"
#p_environment = "Development"
```

```python
%run configuration
```

```python
%run common_functions
```

```python
ope_schema = "OSID INT, Platform STRING, OperatingSystem STRING, \
                SoftwareExtras STRING"
```

```python
df = (
    spark.read
    .schema(ope_schema)
    .option("multiline", "true")
    .json(f"{bronze_folder_path}/{p_file_date}/operationg_system.json")
)
```

##### **2.- Eliminar columnas no deseadas**

```python
df2 = df.drop("OperatingSystem", "SoftwareExtras")
```

##### **3.- Cambiar nombres de columnas**

```python
new_cols = ["os_id", "platform"]

df3 = df2.toDF(*new_cols)
```

##### **4.- AÃ±adir ingestion_date y environment**

```python
from pyspark.sql.functions import lit

df4 = add_ingestion_date(df3) \
        .withColumn("environment", lit(p_environment)) \
        .withColumn("file_date", lit(p_file_date))
```

```python
%%sql
DROP TABLE if EXISTS operationg_system
```

##### **4.- Transformar el archivo a delta y llevar a lh_silver**

```python
df4.write.format("delta").mode("overwrite").saveAsTable("operationg_system")
```
